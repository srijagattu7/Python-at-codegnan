- Methods:The function defination is implemented inside the class those functions called as methods.
- Types of methods
  1. Instance method
  2. class method
  3. static method

- instance Method: The method defination takes first parameter as `self` those methods called as `Instance Method`. This methods works on instance attributes. <br>
  `syntax:`<br>
      `   class class_name:`       <br>
      `       def method_name(self):` <br>
       `          #block code`

- Class method: The method defination starts with `classmethod` decorator and it taks  first parameter as `cls` those methods called as 'Class Methods'. This methods works on class Attributes.  <br>
 `syntax:`<br>
      `   class class_name:`       <br>
      `       @classmethod`      <br>
      `       def method_name(cls, parameters):` <br>
       `          #block code`

- Static Method: The method defination starts with `staticmethod` decorator and it doesn't take first parameter as `self or cls` those methods called as `staticmethods`. This methods works on except instance attributes and class attributes.  <br>
`syntax:`<br>
      `   class class_name:`       <br>
      `       @staticmethod`      <br>
      `       def method_name(parameters):` <br>
       `          #block code`

- Initialization constructor(__init__): The initialization constructor is used for to initilize the inatance attribute data and these constructor takes first parameter as self and these constructor executes whenever an object is created for the class.

```python
#Instance method:
class Student:
    school="ABC school"
    def __init__(self,stdid,name,standed):
        self.stdid=stdid   #instance attributes
        self.name=name    #instance attributes
        self.standed=standed   #instance attributes
    def student_details(self):   #instance attributes
        return f"My self {self.name} and from {self.standed} class"
    #class method
    @classmethod
    def update_school(cls,new_school_name):
        cls.school= new_school_name
        return "School name Updated"
    @staticmethod
    def percentage(marks:list):
        return (sum(marks)/300)*100
        
s1=Student(102,"Aarush",3)
s2=Student(103,"Anshu",2)
print(s1.school, s2.school)
print(s1.student_details())
print(s1.update_school("CODEGNAN school"))
print(s1.school, s2.school)
print(s2.percentage([85,68,96]))
```

### Encapsulation:
 - Encapsulation means, wraping the methods and attributes in single unit that one is class. The main aim of encapsulation is to restrict the direct        access of attribute and methods.
 - This type of restriction can implemented by using 'Access Modifiers"
 - Types of access modifiers:
    * public # starts without underscore
    * protected # starts with single underscore
    * private #starts with dunder score it is private
- In python we don't have any keywords to represent it public, protected and private, but by using underscore(_) cans define.
 1. Public:The Public members(i.e attributes + method) doesnot starts with underscore and this members can access any where in the program.
 2. protected: The protected members starts with single underscore this members can access in inside the class and subclasses also.
 3. private: The Private members start with double underscore and this members can access only inside the class.

    |-----------------|-------------------|-----------|-------------------|<br>
    | Access modifiers   Inside the class    subclass     outside class       <br>
    |-----------------|-------------------|-----------|-------------------|<br>
    |   public               Accepted        Accepted    Accepted           <br>
    |-----------------|-------------------|-----------|-------------------|<br>
    |   protected          Accepted          Accepted                      <br>
    |-----------------|-------------------|-----------|-------------------|<br>
    |   private             Accepted                                        <br>
    |-----------------|-------------------|-----------|-------------------|<br>
- Python doesn't support the high security like java, here we can access the protected members and private members in outside the class by using following ways.
- to access the protected member `object._protected member name`
- To access the private members by using name mangling.
  SYNTAX:
      ` object._classname__private member name`
- The advantage of Encapsulation is can provide the security for data as well as for functionalities.

```python
class Bank:
    def __init__(self, name, account,password,balance):
        self.name=name
        self.account=account
        self.password=password
        self._balance=balance
    
    def __total__users(self):
        return "Total users: 101"
    def get_balance(self):
        return f"Current Balance is: {self._balance}"
    def deposite(self, deposite_amount):
        self._balance += deposite_amount
        return f" {deposite_amount} Deposited Successfully"
u1=Bank("Srija",1001,"srija@123",10000)
print(u1.name)
print(u1.get_balance)
print(u1.deposite(5000))
print(u1.get_balance)
print(u1._balance)   # protected members access in outside the class
print(u1._Bank__total__users())  #name mangling : Syntax: obj._classname__attrinute
```

## INHERITANCE
- In python, a class get properties from another class this processes called as "Inheritance".
- A child class is inherited from parent class.
  TYPES OF INHERITANCES
 1. single Inheritance
 2. multi level Inheritance
 3. multiple Inheritance
 4. hierarchial Inheritance
 5. hybrid Inheritance
# 1. SINGLE INHERITANCE:
   - A child class get properties from single parent class.This is called as "single Inheritance". <br>
     Syntax:<br>
          `class parent_class_name:`<br>
               `#attributes` <br>
               `#methods`     <br>
         `class childclass_name(Parent_class_name):`   <br>
                `#attributes` <br>
               `#methods`      <br>
# 2. MULTI-LEVEL INHERITANCE:
  - A class (i.e child class) get properties from another class(i.e parent class), that class get properties from some another class(i.e Grand parents)
# 3. MULTIPLE INHERITANCE:
  - A single child class get properties from multiple parent classes. This is call as multiple inheritance.
# 4. HIERARCHIAL INHERITANCE:
  - multiple child classes get properties from single parent class.
# 5. HYBRID INHERITANCE:
  - This is a combination of two or more types of inheritences.

```python
#Multiple Inheritance
class Father:
    def __init__(self, father_name):
        self.father_name = father_name
    def show_father(self):
        print("Father Name:", self.father_name)
class Mother:
    def __init__(self, mother_name):
        self.mother_name = mother_name
    def show_mother(self):
        print("Mother Name:", self.mother_name)
class Child(Father, Mother):
    def __init__(self, father_name, mother_name, child_name):
        Father.__init__(self, father_name)
        Mother.__init__(self, mother_name)
        self.child_name = child_name
    def show_child(self):
        print("Child Name:", self.child_name)
p = Child("Rama", "Sita", "LavaKusha")
p.show_father()
p.show_mother()
p.show_child()
```

```python
#Hierarchical Inheritance
class Animal:
    def __init__(self, name):
        self.name = name
    def show(self):
        print("Animal Name:", self.name)
class Dog(Animal):
    def sound(self):
        print("Dog says: Bark")
class Cat(Animal):
    def sound(self):
        print("Cat says: Meow")
class Cow(Animal):
    def sound(self):
        print("Cow says: Moo")
d = Dog("Dog")
c = Cat("Cat")
cw = Cow("Cow")
d.sound()
c.sound()
cw.sound()
```

```python
class Employee:
    def __init__(self, name):
        self.name = name
    def show_employee(self):
        print("Employee Name:", self.name)
class TeamLeader(Employee):
    def __init__(self, name, team):
        super().__init__(name)
        self.team = team
    def show_team(self):
        print("Team:", self.team)
class Manager(TeamLeader):
    def __init__(self, name, team, department):
        super().__init__(name, team)
        self.department = department
    def show_manager(self):
        print("Department:", self.department)
# Creating object
obj = Manager("Srija", "Python Team", "IT")
# Calling methods
obj.show_employee()
obj.show_team()
obj.show_manager()
```

```python
#Polymorphism
#Method Overriding
#Even if you define different numbers of parameters in methods, always child method overrides parent class method.
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
    def details(self):
        return f"My name is {self.name} and I am {self.age} years old"
class Student(Person):
    def __init__(self, stdid, name, age, gender, branch):
        super().__init__(name, age, gender)
        self.stdid = stdid
        self.branch = branch
    def details(self):
        parent_details = super().details()
        return f"{parent_details}. I'm from {self.branch} branch."
s = Student(101, "Krish", 21, "M", "CSC")
print(s.details())
```

```python
#Method overloading
class Math():
    def add(self,a,b):
        return a + b
    def add(self,a,b,c):
        return a+b+c
    def add(self,a,b,c,d):
        return a+b+c+d
m=Math()
print(m.add(1,2,3,4))
#print(m.add(1,2,3))
#print(m.add(1,2))
```

```python
#Method overloading
# Directly python doesn't supports method overloading but, we can achive same functionality through the default arguments or Arbitary methods
class Math():
    def add(self,a,b,c=0,d=0):
        return a+b+c+d
m=Math()
print(m.add(1,2,3,4))
print(m.add(1,2,3))
print(m.add(1,2))
```

```python
class Point:
    def __init__(self,x):
        self.x=x
    def __str__(self): #string magic method
        return f"This is point class and point value is {self.x}"
    def __sub__(self, other):
        return self.x - other.x
    def __mul__(self, other):
        return self.x * other.x
    def __truediv__(self, other):
        return self.x / other.x
    def __le__(self, other):
         return self.x <= other.x
    def __ge__(self, other):
         return self.x >= other.x
    def __mod__(self, other):
         return self.x % other.x

p1=Point(5)
print(p1)
p2=Point(4)
#print(p1.x +p2.x)
p3=p1-p2
print(type(p3))
print(p3)
p3=p1*p2
print(p3)
p3=p1/p2
print(p3)
p3=p1<=p2
print(p3)
p3=p1>=p2
print(p3)
p3=p1%p2
print(p3)
```
