# SET CHARACTERISTICS

```python
# Set unordered collection
#set returns unique elements
s={1,2,5,3,3,'hi','a',3.24,(10,15,20)}
print(s)
#print(s[0]) -> set doesn't supports indexing
```

```python
#list can't be used as set element
s={1,2,5,3,3,'hi','a',3.24,(10,15,20),[3,2]}
print(s)
```

# Set Methods

```python
#Element adding method
#add(ele): It adds the elemnet to set if element not in set, otherwise nothing
s={1,2,4,3,6,5,3}
print(s)
s.add(10)
s.add(5)
print(s)
```

# Element Removing methods
- remove(item):It removes the item if it exists in set otherwise, 'KeyError'
- discard(item):It removes the item if it exists in set otherwise, nothing

```python
s={1,2,3,4,5}
s.remove(5)
print(s)
#s.remove(100) ->KeyError
s.discard(3)
print(s)
s.discard(100) 
print(s)
```

```python
# Sets Operations
#union(set): It returns the all unique elements from all sets
s1={1,2,3,4,}
s2={3,4,8,9}
s3={2,4,5,6}
union=s1.union(s2)
print(union)
union=s1.union(s2,s3)
print(union)
res=s1|s2|s3
print(res)
```

```python
#INTERSECTION
s1={1,2,3,4,}
s2={3,4,8,9}
s3={2,4,5,6}
intersection=s1.intersection(s2)
print(intersection)
intersection=s1.intersection(s2,s3)
print(intersection)
res=s1&s2&s3
print(res)
```

```python
#DIFFERENCE
#difference(set):It returns the a set elements those are not in another sets
s1={1,2,3,4,}
s2={3,4,8,9}
s3={2,4,5,6}
d=s1.difference(s2)
print(d)
d=s1.difference(s2,s3)
print(d)
res=s1-s2-s3
print(res)
print((s1|s2)-s3)
```

```python
#Symmetric DIFFERENCE
#symmetric_difference(set):It returns the non common elements from all sets
s1={1,2,3,4,}
s2={3,4,8,9}
s3={2,4,5,6}
sy_d=s1.symmetric_difference(s2)
print(sy_d)
res=s1^s2^s3
print(res)
```

```python
#issubset(set):
s1={1,2,3,4,}
s2={3,4,8,9}
s3={2,4,5,6}
s=s1.issubset(s2)
print(s)
```

```python
#issuperset(set):
s1={1,2,3}
s2={3,4,1,2,5}
s3={2,4,5,6}
s=s2.issuperset(s1)
print(s)
```

```python
#isdisjoint(set):
s1={1,2,3,4,}
s2={3,4,8,9}
s3={7,5,6}
s=s1.isdisjoint(s3)
print(s)
```
