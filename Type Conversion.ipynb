```python
#string to int,bool,list,tuple,set
s='123456'
print(int(s))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
#string to int,bool,list,tuple,set
s='123456.1205'
print(int(float(s)))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
#list to str,tuple,set
x=[1,12,15,18]
str_val=str(x)
print(str_val)
print(tuple(x))
print(set(x))
```

```python
#tuple to str,list,set
x=(1,12,15,18)
print(str(x))
print(list(x))
print(set(x))
```

```python
#set to str,list,tuple
s={1,2,5,3,3,'hi','a',3.24,(10,15,20)}
print(str(s))
print(list(s))
print(tuple(s))
```

```python
#dict to str,list,set,tuple
d={"Name":"Gattu","Course": "Python","Batch":"57", "age":"21"}
print(str(d))
print(set(d))
print(list(d))
print(tuple(d))
```
