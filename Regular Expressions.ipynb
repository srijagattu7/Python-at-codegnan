```python
import re
```

```python
#match
string ="The sky is blue and ocean is blue"
res=re.match("The", string)
print(res)
print(res.span())#it returns the matched pattern positions
print(res.group()) #It returns the Matched result
```

```python
#match(pat,string): It checks the whether the string is starts with pattern or not.
try:
    string ="The sky is blue and ocean is blue"
    res=re.match("sky", string)
    print(res)
    print(res.span())#it returns the matched pattern positions
    print(res.group()) #It returns the Matched result
except:
    print("There is no Matched pattern")
```

```python
#search(pat,string): It checks the entire string and it return the first matched pattern.
try:
    string ="The sky is blue and ocean is blue"
    res=re.search("sky", string)
    print(res)
    print(res.span())#it returns the matched pattern positions
    print(res.group()) #It returns the Matched result
except:
    print("There is no Matched pattern")
```

```python
#findall(pat,string):It returns the all matched patterns as list
try:
    string ="The sky is blue and ocean is blue"
    res=re.findall("[spe]", string)
    print(res)
except:
    print("There is no Matched pattern")
```

```python
#split(pat,string):It splits the string wherever the pattern is matched.
try:
    string ="The sky is blue and ocean is blue"
    res=re.split("blue", string)
    print(res)
except:
    print("There is no Matched pattern")
```

```python
try:
    string ="The sky is blue1 and ocean is blue"
    res=re.split("1", string)
    print(res)
except:
    print("There is no Matched pattern")
```

```python
#sub(old_pat,new_patstring):It splits the string wherever the pattern is matched.
try:
    string ="The sky is blue and ocean is blue."
    res=re.sub("blue", "Red", string)
    print(res)
except:
    print("There is no Matched pattern")
```

```python
#validate mobile number
try:
    mobile=input()
    res=re.fullmatch("^[6-9][0-9]{9}",mobile)
    print(res)
    if res:
        print(f"{mobile} is valid")
    else:
        print(f" {mobile} is Invalid")
except:
    print("There is no matched pattern")
```

```python
#validate Pan number
try:
    pan=input()
    res=re.fullmatch("^[A-Z]{5}[0-9]{4}[A-z]{$}",pan)
    print(res)
    if res:
        print(f"{pan} is valid")
    else:
        print(f" {pan} is Invalid")
except:
    print("There is no matched pattern")
```

```python
#validate Aadhar number
try:
    aadhar=input()
    res=re.fullmatch("^[2-9][0-9]{3}\s?[0-9]{4}\s?[0-9]{4}",aadhar)
    print(res)
    if res:
        print(f"{aadhar} is valid")
    else:
        print(f" {aadhar} is Invalid")
except:
    print("There is no matched pattern")
```

```python
#validate Email 
try:
    email=input()
    res=re.fullmatch("^[a-z0-9][a-z0-9.]+\@[a-z0-9]+\.[a-z]{2,3}$",email)
    print(res)
    if res:
        print(f"{email} is valid")
    else:
        print(f" {email} is Invalid")
except:
    print("There is no matched pattern")
```
