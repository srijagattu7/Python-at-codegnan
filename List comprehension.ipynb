```python
#generate 1-n numbers using list comprehension
n=int(input())
res=[x for x in range(1,n+1)]
print(res)
```

```python
#first n even numbers using list comprehension
n=int(input())
res=[x*2 for x in range(0,n)]
print(res)
```

```python
#find the first n cubic numbers
n=int(input())
res=[x**3 for x in range(1,n+1)]
print(res)
```

```python
#find odd numbers between n1 and n2
n1,n2=map(int, input().split())
res=[x for x in range(n1,n2+1) if x%2==1]
print(res)
```

```python
#In given arr of integers perform following operations
#if even-find square of a number
#if odd-find squareroot value(upto 2 decimal places)
arr=list(map(int,input().split()))
res=[x**2 if x%2==0 else round(x**0.5,2) for x in arr ]
print(res)
```

```python
#find the 2 dice combination values
res=[(x,y) for x in range(1,7) for y in range(1,7)]
print(res)
```

```python
#Sum of 2 dice face value is divisible by 3
res=[(x,y) for x in range(1,7) for y in range(1,7) if (x+y)%3==0]
print(res)
```

```python
res=[(x,y) for x in range(1,7) for y in range(1,7) if x==4 or y==4]
print(res)
```

```python
#count the number of vowels exists in string
s="srija"
res=[char for char in s.lower() if char in "aeiou"]
print(res)
print(len(res))
```

```python
s="srija"
res=sum([1 for char in s.lower() if char in "aeiou"])
print(res)
```

```python
#set comprehension
#1 to n numbers
n =int(input())
res={x for x in range(1,n+1)}
print(type(res))
print(res)
```

```python
#dictionary comprehension
n =int(input())
res={x:x**2 for x in range(1,n+1)}
print(type(res))
print(res)
```

```python
# TUPLE comprehension
n =int(input())
res=tuple(x for x in range(1,n+1))
print(type(res))
print(res)
```

```python
# Whether the person is eligible for vote or not
age=int(input())
if age>=18:
    print("Eligible")
else:
    print("Not Eligible")
```
