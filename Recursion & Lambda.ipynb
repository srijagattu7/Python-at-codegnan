```python
def factorial(x):
    if x==0 or x==1:
        return 1
    return x*factorial(x-1)
print(factorial(5))
```

```python
string=input()
count={}
for char in string:
    if char in count:
        count[char]+=1
    else:
        count[char]=1
print(count)
```
