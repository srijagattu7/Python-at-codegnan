## STRING METHODS
-Case conversion Methods
    -upper():It converts entire string into upper case
    -lower():It converts entire string into Lower case
    -title() :It converts the first letter of every word into upper and remaining lower
    -swapcase(): It coverts Upper case letter into lower case and lower to upper
    -capitalize(): It converts First character into upper case and remaining into lower case

```python
s="We are from Warangal"
print(s.upper())
print(s.lower())
print(s.title())
print(s.swapcase())
print(s.capitalize())
```

- Remove white space with pattern at ends of string
      - strip(): By default it removes the white space at both ends of string
      - lstrip(): It removes the white space at left side of string
      - rstrip(): It removes the white space at right side of string

```python
s="     Gattu       "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s="@@@@Ga@@ttu@@@"
print(s)
print(s.strip("@"))
print(s.lstrip("@"))
print(s.rstrip("@"))
```

```python
# split(): It splites the entire string into list of substring based on specified seperator
# By default it splits at space
s="The sky is blue and Ocean is Blue"
print(s.split())
print(s.split("is"))
print(s.split('e'))
```

```python
#'sep'.join(iterable): It joins the all iterable elements with seperator
w=['The', 'sky', 'is', 'blue', 'and', 'Ocean', 'is', 'Blue']
print("".join(w))
print(" ".join(w))
print("_".join(w))
```

# Substring Index Finding
- index(substring):If substrings is exists in string, it returns the first occurence of substring
   position, otherwise 'ValueError'
- find(substring):If substrings is exists in string, it returns the first occurence of substring
   position, otherwise '-1'

```python
s="The sky is blue and Ocean is Blue"
print(s.index('is'))
print(s.index('z'))
```

```python
s="The sky is blue and Ocean is Blue"
print(s.find('is'))
print(s.find('z'))
print(s.find('is', 10)) # Its starts checking from 10th position
print(s.find('e', 10))
print(s.find('e', -5))
```

```python
#count(substring): It returns the  non-overlapping occurences of substing
s="ABCBCBCbbaabb"
print(s.count("B"))
print(s.count("BCB"))
lower_string=s.lower()
print(lower_string)
print(lower_string.count('b'))
print(s.upper().count('B'))
print(s.count('Z'))
```

```python
#replace(old_substring,new_substring): It replace the old sub-string with new sub-string in string
s="The sky is blue and Ocean is Blue"
print(s.replace("blue", "Red"))
print(s.replace('e','E'))
print(s.replace('e','E',2))#it replace only first 2 occurance
```

## STRING CHECKING METHODS
*String checking methods returns the Boolean value as result*
- isupper():It return True, If entire string is in upper case
- islower():It return True, If entire string is in lower case
- isalpha():It return True, If entire string contains alphabets
- isdigit():It return True, If entire string contains digits
- isalnum():It return True, If entire string contains either alphabets or digits
- startswith(substring): It return True, If string starts with substring
- endswith(substring): It returns True, if string ends with substring

```python
print("CODE".isupper())
print("CODE".islower())
print('code'.islower())
print('1234'.isdigit())
print('123abc'.isdigit())
print('Code'.isalpha())
print('Code123'.isalnum())
print("code".startswith('c'))
```
