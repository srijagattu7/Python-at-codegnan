```python
#Nested If
x=int(input())
if x>=0:
    if x%2==0:
        print(x,"is +ve even")
    else:
        print(x,"is +ve odd")
else:
    if x%2==0:
        print(x,"is -ve even")
    else:
        print(x,"is -ve odd")
```

```python
x=input().lower()
v=0
c=0
for a in x:
    if a in 'aeiou':
        v+=1
    else:
        c+=1
print("vowels", v)
print("consonants", c)
```
