# Dictionary Characteristics

```python
#Dictionary ordered collection
details={"Name":"Gattu","Course": "Python","Batch":"57", "age":"21"}
print(details['Course'])
#print(details['Location']) ->keyError:'Location'

#updating dictionary values
details['age']=22
print(details['age'])
#adding new items
details['location']="Hyderabad"
print(details)
```

```python
#Dictionary keys are unique
details={'Name': 'Gattu', 'Course': 'Python', 'Batch': '57', 'age': 22, 'Marks': '[10,20,30]'}
print(details)
print(details[0])
```

## DICTIONARY METHODS
- get(key,default_value:Optional):It returns the value if key is exists in dictionary,otherwise it returns default value

```python
#Dictionary keys are unique
details={'Name': 'Gattu', 'Course': 'Python', 'Batch': '57', 'age': 22, 'Location': 'Hyderabad'}
print(details['Name'])
#print(details['Marks']) ->This one raises keyError
print(details.get('Name'))#get method bydefault it returns None if key not exists in
print(details.get('marks', "Marks not available")) #it returns default value because,
```

- update(new_dictionary):This method updates the values if new_dictionary keys already exists in Dictionary, otherwise it adds new items to dictionary

```python
details={'Name': 'Gattu', 'Course': 'Python', 'Batch': '57', 'age': 22,}
details.update({'Batch':58,'course':'Java','Location':'HYD'})
print(details)
```

- keys():This method returns the list dictionary keys
- values(): This method returns the list dictionary values
- items(): This method return the list of tuple of key-value pairs

```python
details={'Name': 'Gattu', 'Course': 'Python', 'Batch': 58, 'age': 22, 'course': 'Java', 'Location': 'HYD'}
keys=details.keys()
print(type(keys))
print(keys)
print(details.values())
print(details.items())
```

- popitem() It return and removes the last item as tuple
- pop(key): This method removes the specified key item and return the specified key's value, otherwise KeyError

```python
details={'Name': 'Gattu', 'Course': 'Python', 'Batch': 58, 'age': 22, 'course': 'Java', 'Location': 'HYD'}
item=details.popitem()
print(item)
print(details)
val=details.pop('Name')
print(val)
print(details)
```
