# INPUT FORMATTING

```python
# reading integer from user
n=int(input("Enter a number:"))
print(type(n))
print(n)
```

```python
#Float values from user
f=float(input("Enter a number:"))
print(type(f))
print(f)
```

```python
#Reading Complex from user
c=complex(int(input()),int(input()))
print(type(c))
print(c)
```

```python
#reading boolean values
b=bool(input())
print(type(b))
print(b)
```

```python
#reading boolean as integer
b=bool(int(input()))
print(type(b))
print(b)
```

```python
#reading string value from user
b=str(input())
print(type(b))
print(b)
```

```python
#reading list value from user
b=list(input())
print(type(b))
print(b)
```

```python
#reading list value from user
b=map(int,input().split())
print(b)
```

```python
#reading list value from user
b=list(map(int,input().split()))
print(b)
```

```python
#reading list of string from user
l=input().split()
print(l)
```

```python
l=str(input().split())
print(l)
```

```python
b=tuple(map(int,input().split()))
print(b)
```

```python
b=set(map(int,input().split()))
print(b)
```

```python
#eval():This function evaluates the string type expression
val=eval(input())
print(type(val))
print(val)
```

# Output Formatting

```python
#old style formatting using comma
s='srija'
age=20
print("my name is",s,"and My age is",age)
```

```python
# Access Specifiers
s='srija'
age=20
print("my name is %s and My age is %d"%(s,age))
```

```python
s='srija'
age=20
print("my name is %s and My age is %d"%(age,s))
```

```python
#Dot Format
s='srija'
age=20
print("my name is {} and My age is {}".format(s,age))
print("my name is {} and My age is {}".format(age,s))
```

```python
#f-Format
name='srija'
age=20
print(f"my name is {name} and My age is {age}")
print(f"my name is {age} and My age is {name}")
```
