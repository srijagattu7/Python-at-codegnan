```python
#Types or arguments
#1. Positional Arguments: Arguments are assign to parameters as per the sequential order
#function defination block
def sub(x:int,y:int)->int:
    z=x-y
    return z
#function calling block
a,b=20,10
print(sub(a,b))
print(sub(b,a))
```

```python
#Keyword Arguments: Arguments are explicitly passed with parameter names
#function defination block
def sub(x:int,y:int)->int:
    z=x-y
    return z
#function calling block
a,b=20,10
print(sub(x=a,y=b))
print(sub(y=b,x=a))
```

```python
def add(x:int,y:int,w:int)->int:
    z=x+y+w
    return z
#function calling block
a,b,c=20,10,12
print(add(x=a,y=b,w=c))
print(add(y=b,w=c,x=a))
```

```python
def add(x:int,y:int,w:int,v=int)->int:
    z=x+y+w+v
    return z
#function calling block
a,b,c,d=map(int,input().split())
print(add(x=a,y=b,w=c,v=d))
print(add(y=b,w=c,v=d,x=a))
```

```python
def multi(x,y):
    z=x*y
    return z
a,b=map(int,input().split())
print(multi(a,b))
```

```python
#3.Default arguments:The default arguments are optinal arguments,if you didn't pass value 
#               for default argument, in that case it assigns default value otherwise, it assigns new argument value.
def add(w:int,x:int,y:int=0,z:int=0)->int:
    r=w+x+y+z
    return r
#function calling block
a,b,c,d=map(int,input().split())
print(add(a,b))
print(add(a,b,c))
print(add(a,b,c,d))
```

```python
def add(w:int,x:int,y:int=1,z:int=1)->int:
    r=w*x*y*z
    return r
#function calling block
a,b,c,d=map(int,input().split())
print(add(a,b))
print(add(a,b,c))
print(add(a,b,c,d))
```

```python
#Variable length Positional arguments
def add(x,y=0,*args):
    print(type(args))
    #print(args)
    s=0
    for val in args:
        s += val
    return s
print(add(10,20,1,2,3,4,5))
print(add(10,20,1,2,3,4,5,6,7,8,9))
```

```python
#variable length Keyword Arguments
def add(x,y=0,*args,**kwargs):
    print(type(kwargs))
    print(kwargs)
    print(kwargs.values())
    print(kwargs.keys())
print(add(1,2,3,4,5,name='srija',age=21,course='PFS'))
```

```python
#In python every type function calling block shares the arguments reference to parameters 
#   if the arguments are immutable data types,then behave likes pass by value and if arguments are mutable 
#   datatypes(i.e list,set,dict) then behave likes pass by reference
def fun(x):
    print(id(x))
    x+=2
    print("after adding 2")
    print(id(x))
    print(x)
a=5
print("Before function Execution 'a' value:",a)
fun(a)
print("After function Execution 'a' value:",a)
```
