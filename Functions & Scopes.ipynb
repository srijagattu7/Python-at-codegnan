```python
#Types of functions based on parameters and return value
#4. With parameters and with return
#function defination block
def addition(x:int, y:int)->int:
    c=x+y
    return c
#function calling block
a,b=10,20
res=addition(a,b)
print(res)
print(addition(4,5))
```

```python
#3. With parameters and without return
#By default function returns None value, When function not returning any value
def addition(x:int, y:int)->None:
    c=x+y
    print(f"This is from Function defination block:{c}")
#function calling block
a,b=10,20
res=addition(a,b)
print(f"This is from Function calling block:{res}")
```

```python
#2. without parameters and with return
def addition()->int:
    x=10
    y=20
    z=x+y
    return z
#function calling block

res=addition()
print(f"This is from Function calling block:{res}")
```

# SCOPE OF VARIABLES


```python
#Scopes of variables
#1. Local variable
#2. Enclosing variable
#3. Global scope
#4. built-in scope
x=10  #Global scope variable
def outer():
    y=20   # enclosing scope variable
    def inner():
        z=30   #local scope
        print(z,y,x)   #Built-in scope function
    inner()
outer()
```

```python
#using global scope variable in local scope
x=100
def outer():
    y=x+1
    print("Local scope variable value:",y)
outer()
print("Global scope variable value:",x)
```

```python
#updating the global scope variable value in local scope
#can't update global scope variable in local scope directly.
x=100
def outer():
    x=x+1
    print("Local scope variable value:",x)
outer()
print("Global scope variable value:",x)
```

```python
#updating the global scope variable value in local scope by using global keyword
x=100
def outer():
    global x
    x=x+1
    print("Local scope variable value:",x)
outer()
print("Global scope variable value:",x)
```

```python
#updating the enclosing scope variable value in local scope
def outer():
    y=20  
    def inner():
        z=30
        y=y+z
        print(y,z)
    inner()
outer()
```

```python
#updating the enclosing scope variable value in local scope by using non-local
def outer():
    y=20  
    def inner():
        nonlocal y
        z=30
        y=y+z
        print(y,z)
    inner()
    print("Enclosing scope variable value:",y)
outer()
```
