```python
s1="CODE"
s2="GNAN"
n=len(s1)
s=""
for i in range(n):
    diff=(abs(ord(s1[i])-ord(s2[i])))
    if diff==0:
        s+=s1[i]
    else:
        s+=chr(64+diff)
print(s)
```

```python
s1="CODE"
s2="GNAN"
n=len(s1)
s=""
b=""
for i in range(n):
    if s1[i]!=s2[i]:
        if s1[i]>s2[i]:
            s+=s2[i]
            b+=s1[i]
        else :
            b+=s2[i]
            s+=s1[i]
print(s)
print(b)
```

```python
#map(fun,iterable): It applies the functionality on each element in iterable
#filter(fun,iterable): It filters the iterables, it returns the values which is not None
#find the square of list of integers
def square(x):
    return x**2
lst=[1,2,3,4,5]
print(list(map(square,lst)))
```

```python
#Filter the even numbers from list of integers
def even(x):
    if x%2==0:
        return 1
arr=[1,2,3,4,5,6,7]
print(list(filter(even,arr)))
print(list(map(even,arr)))
```

```python
def square(t):
    x,y=t
    return x+y
lst=[(1,2),(3,4)]
print(list(map(square,lst)))
```

```python
#zip(iterable1, iterable2,.......,iterableN):
#It returns list of tuples, the tuple is generated as each position elements from all iterables.
l1=['a','b','c']
l2=[1,2,3,4,5]
print(list(zip(l1,l2)))
```

```python
#enumerate(itarable): It returns the list of tuple of index and value
arr=[10,20,30]
print(list(enumerate(arr)))
```

```python
#find the target element position in list
arr=[10,20,30,40,50]
target=40
for ind, val in enumerate(arr):
    if target == val:
        print(ind)
        break
else:
    print(-1)
```

```python
#any(iterable): It returns True when any one iterable element's boolean value is TRUE.
arr=[1,2,4, 0,0,0]
print(any(arr))
print(any([0,0,0,0,0]))
#all(iterable):It returns True when all iterable element's boolean value is TRUE.
print(all([1,2,3,4]))
print(all([True,False,True]))
print(all([False,False,False]))
```

```python
#isinstance(obj,class_name):It checks whether the object is belongs to particular class or not.
print(isinstance(10,int))
print(isinstance(12.25,int))
```

```python
#sorted():it returns the new sorted object, by default it returns the desending order
#reverse():it returns the reverse of an object value
arr=[10,2,34,35,3,5,6]
res=sorted(arr,reverse=True)
print(res)
print(arr)
print(list(reversed(arr)))
print("".join(list(reversed("Hello"))))
```
