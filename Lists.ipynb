## List Characteristics

```python
#Define list
lst=[1,5,2.6,"Hi","Hello",1,4,5]
print(lst)
```

```python
#list allows indexing and slicing
lst=[1,5,2.6,"Hi","Hello",1,4,5]
print(lst[5])
print(lst[-5])
print(lst[12])
```

```python
lst=[5,10,15,20,25,30,35]
print(lst[:5:-1])
print(lst[:500])
print(lst[::-1])
```

```python
#Empty list creation
l1=[]
l2=list()
print(type(l1),type(l2))
```

```python
#LIST OPERATIONS
#1.list concatination (using + operator)
#2.list repitation  (using * operator)
l1=[1,5,8]
l2=[6,7,9]
print(l1+l2)
print(l1+l1)
print(l2*3)
print([0]*5)
```

## LIST METHODS
- Item adding methods
   - append(item):It adds item at end of list
   - extend(iterable):It adds the all iterator elements at end of list
   - insert(pos, item): It adds item at specified position

```python
lst=[15,25,35,45]
lst.append(50)
print(lst)
lst.extend([5,6,8])
print(lst)
lst.append([50,70])
print(lst)
lst.insert(1,80)
print(lst)
```

# ITEM REMOVING METHODS
- remove(items):It removes the first occurance item from the list if it is exists in list otherwise 'ValueError'
- pop(pos):It remove and return the specified position item if position is passed, otherwise it remove and returns the last item from list
- clear():It removes all elements from list


```python
lst=[2,4,6,8,4]
lst.remove(4)
print(lst)
```

```python
lst=[2,4,6,8,4]
val=lst.pop()
print(val)
print(lst)
val2=lst.pop(0)# It removes the 0th position
print(val2)
print(lst)
lst.clear()
print(lst)
```

```python
#Index and count()
# index(item): If item is present in list it returns first occurance the Index otherwise 'ValueError'
# count(item): It return the count of total occurnces of item
lst=[10,20,30,40,30]
ind=lst.index(20)
print(ind)
print(lst.count(30))
```

```python
#sorting items
#sort(reverse=False): By default it sorts the original list in into ascending order. 
 #                            if reverse is true it sorts in decending order
lst=[1,2,5,-23,5,-25,-5,25]
lst.sort()
print(lst)
lst.sort(reverse=True)
print(lst)
```

```python
lst=['Hi','Hello','Aarush']
lst.sort()
print(lst)
lst.sort(reverse=True)
print(lst)
```

```python
# Builtin functions - len(), min(), max(), sum()
lst=[2,5,8,9]
print(len(lst),min(lst),max(lst),sum(lst))
print("Average:",sum(lst)/len(lst))
```
