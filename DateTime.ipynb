```python
from datetime import datetime
current_timestamp=datetime.now()# to get the current timestamp
print(current_timestamp)
today=current_timestamp.date()  # to extract date from current timestamp
print(today)
current_time=current_timestamp.time()  # to extract time from current timestamp
print(current_time)
#extract day, month, and year
day=today.day
month=today.month
year=today.year
print(day, month, year)
#Extract hour, minute, and seconds from time
print(current_time.hour)
print(current_time.minute)
print(current_time.second)
#change date format
print(today.strftime("%A %d/%m/%Y"))
print(current_time.strftime("%I:%M %p"))
```

```python
#Timer implementation
import time
from IPython.display import clear_output
hour_counter=minute_counter=second_counter=0
while True:
    print(f"{hour_counter}:{minute_counter}:{second_counter}")
    time.sleep(1) #It pause the execution for 1 sec
    second_counter+=1
    if second_counter==60:
        minute_counter+=1
        second_counter=0 #reset seconds counter to 0
        if minute_counter==60:
            hour_counter+=1
            minute_counter=0  #reset minute counter to 0
            if hour_counter==24:
                hour_counter=0
    clear_output(wait=True)
```
