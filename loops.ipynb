```python
#Break
lst=list(map(int,input().split()))
target=int(input())
n=len(lst)
for i in range(n):
    if lst[i]==target:
        print(i)
        break
else:
    print(-1)
```

```python
n=int(input())
f=list()
for i in range(1,n):
    if n%i==0:
        f.append(i)
if len(f) == 2:
    print(f"{n} is prime")
else:
    print(f"{n} is not prime")
```

```python
#Prime number
n=int(input())
for i in range(2,n//2+1):
    if n%i==0:
        print(f"{n} not prime")
        break
else:
    print(f"{n} prime")
```
